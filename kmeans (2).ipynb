```python
data='/kaggle/input/kmeanssss/leaf.csv'
df = pd.read_csv(data)
df.head()
```

```python
km = KMeans(n_clusters=10)
y_predicted = km.fit_predict(df[['Class (Species)','Specimen Number']])
y_predicted
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
df6 = df[df.cluster==5]
df7 = df[df.cluster==6]
df8 = df[df.cluster==7]
df9 = df[df.cluster==8]
df10 = df[df.cluster==9]
plt.scatter(df1['Class (Species)'],df1['Specimen Number'],color='green')
plt.scatter(df2['Class (Species)'],df2['Specimen Number'],color='red')
plt.scatter(df3['Class (Species)'],df3['Specimen Number'],color='blue')
plt.scatter(df4['Class (Species)'],df4['Specimen Number'],color='white')
plt.scatter(df5['Class (Species)'],df5['Specimen Number'],color='yellow')
plt.scatter(df6['Class (Species)'],df6['Specimen Number'],color='pink')
plt.scatter(df7['Class (Species)'],df7['Specimen Number'],color='orange')
plt.scatter(df8['Class (Species)'],df8['Specimen Number'],color='cyan')
plt.scatter(df9['Class (Species)'],df9['Specimen Number'],color='magenta')
plt.scatter(df10['Class (Species)'],df10['Specimen Number'],color='gray')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Class (Species)')
plt.ylabel('Specimen Number')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Class (Species)']])
df['Class (Species)'] = scaler.transform(df[['Class (Species)']])

scaler.fit(df[['Specimen Number']])
df['Specimen Number'] = scaler.transform(df[['Specimen Number']])
df.head()
```

```python
plt.scatter(df['Class (Species)'],df['Specimen Number'])

km = KMeans(n_clusters=10)
y_predicted = km.fit_predict(df[['Class (Species)','Specimen Number']])
y_predicted

df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
df6 = df[df.cluster==5]
df7 = df[df.cluster==6]
df8 = df[df.cluster==7]
df9 = df[df.cluster==8]
df10 = df[df.cluster==9]
plt.scatter(df1['Class (Species)'],df1['Specimen Number'],color='green')
plt.scatter(df2['Class (Species)'],df2['Specimen Number'],color='red')
plt.scatter(df3['Class (Species)'],df3['Specimen Number'],color='blue')
plt.scatter(df4['Class (Species)'],df4['Specimen Number'],color='white')
plt.scatter(df5['Class (Species)'],df5['Specimen Number'],color='yellow')
plt.scatter(df6['Class (Species)'],df6['Specimen Number'],color='pink')
plt.scatter(df7['Class (Species)'],df7['Specimen Number'],color='orange')
plt.scatter(df8['Class (Species)'],df8['Specimen Number'],color='cyan')
plt.scatter(df9['Class (Species)'],df9['Specimen Number'],color='magenta')
plt.scatter(df10['Class (Species)'],df10['Specimen Number'],color='gray')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Class (Species)','Specimen Number']])
    sse.append(km.inertia_)

plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
