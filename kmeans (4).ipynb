```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_stand=sc.fit_transform(df)
data_stand
```

```python
data_stand=pd.DataFrame(data_stand,columns=['Class (Species)','Specimen Number','Eccentricity','Aspect Ratio','Elongation','Solidity','Stochastic Convexity','Isoperimetric Factor','Maximal Indentation Depth','Lobedness','Average Intensity','Average Contrast','Smoothness','Third moment','Uniformity','Entropy'])
data_stand
```

```python
df.plot(kind='scatter',x='Class (Species)',y='Specimen Number')
plt.show()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Class (Species)','Specimen Number']])
    sse.append(km.inertia_)

plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
    
```

```python
km4=KMeans(n_clusters=4)
km4=km4.fit(df)
print(km4.labels_)
```

```python
labels=km4.labels_
labels=pd.DataFrame(labels,columns=['clusters'])
data=pd.concat([df,labels],axis=1)
data.sort_values('clusters',ascending=True)
```

```python
print(km4.labels_)
data.groupby('clusters').size()
```

```python
data.groupby('clusters').describe()
```

```python
centroids=km4.cluster_centers_
centroids=pd.DataFrame(centroids,columns=['Centroid_Class(Species)','Centroid_SpecimenNumber','Centroid_Eccentricity','Centroid_AspectRatio','Centroid_Elongation','Centroid_Solidity','Centroid_StochasticConvexity',
       'Centroid_IsoperimetricFactor','Centroid_MaximalIndentationDepth','Centroid_Lobedness','Centroid_AverageIntensity','Centroid_AverageContrast','Centroid_Smoothness','Centroid_Thirdmoment','Centroid_Uniformity','Centroid_Entropy'])
centroids
```

```python
import seaborn as sns 
s1=sns.scatterplot(data,x='Specimen Number',y='Class (Species)',hue='cluster')
centroids.plot(ax=s1,kind='scatter',x='Centroid_Class(Species)',y='Centroid_SpecimenNumber',color='red',fontsize=14)
```
